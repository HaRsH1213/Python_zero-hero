```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_excel('linechart.xlsx')
df.head()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df['Quarter'],df['Fridge'],color='blue',label='Fridge')
plt.plot(df['Quarter'],df['Dishwasher'], color='red',label='Dishwasher' )
plt.plot(df['Quarter'],df['Washing Machine'],color='orange',label='Washing Machine')

plt.title('Products Sales')
plt.xlabel('Financial Quater')
plt.ylabel('Revenue in (mln $)')
plt.legend()

plt.show()
```

```python
total_revenue=df[['Fridge','Dishwasher','Washing Machine']].sum()
total_revenue
```

```python
plt.pie(total_revenue,labels=['Fridge','Dishwasher','Washing Machine'],autopct='%1.1f%%',explode=(0.1,0,0),shadow=True)
plt.show()
```

```python
df.plot(kind='bar',x='Quarter')
plt.xticks(rotation=45)
plt.show()
```
